# **Task 1**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

B = np.array([[21, 11],[12, 3]])

plt.xlim(0,30)
plt.ylim(0,30)

plt.quiver([0,0],[0,0], B[:,0], B[:,1], 
           angles='xy', scale_units='xy',scale=1, 
           color=['darkturquoise','purple'])## use column spaces

plt.title("Laboratory 5 - Task 1")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()          
plt.show()
```

**General Linear Equation**

$$B = \begin{bmatrix} 21 \ 12\\11 \ 3 \\\end{bmatrix}\\ or \\
B = \left\{
    \begin{array}\\
        21x + 11y \\ 
        12x + 3y
    \end{array}
\right. $$

# **Task 2**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

M = np.array([[0, 7, 14], [5, 5, 5], [5, 7, -13]])

fig = plt.figure(figsize=(10,8))
M1 = fig.gca(projection='3d')

M1.set_xlim([0, 15])
M1.set_ylim([0, 15])
M1.set_zlim([0, 15])
origin = (0, 0, 5)

M1.quiver(origin, origin, origin, M[:, 0], M[:, 1], M[:, 2],
          arrow_length_ratio=0.10, 
          color=['gold', 'yellow', 'gold'])

plt.title("Laboratory 5 - Task 2")
plt.grid() 
plt.show()
```

**General Linear Equation**

$$M = \begin{bmatrix} 0 \ 5 \ 5\\ 7 \ 5 \ 7\\ 14 \ 5 \ -13 \\\end{bmatrix}\\ or \\
M = \left\{
    \begin{array}\\
        7y + 14z\\ 
        5x + 5y + 5z\\
        5x + 7y + -13z\\
    \end{array}
\right. $$
